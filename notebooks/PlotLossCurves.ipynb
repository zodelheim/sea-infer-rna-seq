```python
colors = {
    "chr_aXY": "green", 
    "autosomes": "red", 
    "chr_aX": "blue", 
    "chr_aY": "orange"
}
organ_names = {
    "BRAIN0": "BRAIN0",
    "HEART": "HEART",
    "BRAIN1": "BRAIN1",
    "None": "BLOOD",
    "CAGE.HEART": "CAGE.HEART",
}

for organ in ["BRAIN0", "HEART", "BRAIN1", "None"]:
    fig, axs = plt.subplots(2,2)
    for ax, sex_chromosome in zip(axs.ravel(), ["chr_aXY", "autosomes", "chr_aX", "chr_aY"]):
        with open(PROJ_ROOT/'logs'/f"{organ}{sex_chromosome}_model_log.json", 'r') as file:
            data = json.load(file)

            for dd in data:
                ax.plot(dd, color=colors[sex_chromosome])
            ax.set_title(sex_chromosome)
        # ax.title = f"{organ}: {sex_chromosome}"
        # ax.set_title(f"{organ}: {sex_chromosome}")
    plt.tight_layout()
    fig.suptitle(organ_names[organ])
    plt.show()
            
```
